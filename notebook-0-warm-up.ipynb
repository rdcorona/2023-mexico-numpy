# Lists vs numpy arrays
**So why using arrays?**

Speed matters: We compare math package functions on lists, with a list comprehension, and with numpy universal functions (ufunc)

```python
import math
import timeit
```

# Lists vs numpy arrays
**So why using arrays?**

Speed matters: We compare math package functions on lists, **with a list comprehension**, and with numpy universal functions (ufunc)

```python
# List of numbers
numbers = [0,1,2,3,4,5,6,7,8]

# Compute the square roots and measure the execution time
result = sum([timeit.timeit(lambda: math.sqrt(num), number=1000000) for num in numbers])

# Print the results
print(f"Execution time: {result:.6f} seconds")
```

# Lists vs numpy arrays
**So why using arrays?**

Speed matters: We compare math package functions on lists, with a list comprehension, and **with numpy universal functions (ufunc)**

```python
# List of numbers
numbers = np.array(numbers)

# Compute the square roots and measure the execution time
result = timeit.timeit(lambda: np.sqrt(numbers), number=1000000)

# Print the results
print(f"Execution time: {result:.6f} seconds")
```
